```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from mpl_toolkits.mplot3d import Axes3D

class PhysicalConstants:
    """Physical and electrochemical constants"""
    def __init__(self):
        # Fundamental constants (unchanged)
        self.F = 96485.33212  # Faraday constant (C/mol)
        self.R = 8.31446261815324  # Universal gas constant (J/mol⋅K)
        self.k_b = 1.380649e-23  # Boltzmann constant (J/K)
        self.e = 1.602176634e-19  # Elementary charge (C)
        self.N_A = 6.02214076e23  # Avogadro constant (1/mol)

class MaterialProperties:
    """Properties of common membrane and electrode materials"""
    def __init__(self):
        # Nafion membrane properties (adjusted)
        self.lambda_base = 14  # Base water content (mol H2O/mol SO3-)
        self.EW = 1100  # Equivalent weight (g/mol)
        self.rho_dry = 2.0  # Dry density (g/cm³)
        
        # Carbon paper electrode properties
        self.porosity = 0.78  # Typical porosity
        self.tortuosity = 2.5  # Typical tortuosity
        self.fiber_diameter = 7.5e-6  # Carbon fiber diameter (m)
        
        # Bipolar plate properties (adjusted)
        self.bp_conductivity = 1000  # Electrical conductivity (S/m)  
        self.contact_resistance = 200  # Contact resistance (mΩ⋅cm²) 


class EnhancedOhmicModel(PhysicalConstants, MaterialProperties):
    def __init__(self, t_mem=50e-6, T_cell=353.15):
        PhysicalConstants.__init__(self)
        MaterialProperties.__init__(self)
        
        self.t_mem = t_mem  # Membrane thickness (m)
        self.T_cell = T_cell  # Cell temperature (K)
        
    def water_uptake(self, RH=0.99, T=353.15):
        """Calculate membrane water uptake based on RH and temperature
        
        Args:
            RH (float): Relative humidity (0-1)
            T (float): Temperature (K)
        Returns:
            float: Water content (λ)
        """
        # Base lambda at reference temperature (30°C)
        lambda_ref = 0.043 + 17.81 * RH - 39.85 * RH**2 + 36.0 * RH**3
        
        # Temperature correction factor
        # Modest increase with temperature (~2-3% per 10K above 303.15K)
        temp_factor = 1 + 0.0026 * (T - 303.15)
        
        return lambda_ref * temp_factor
    
    def membrane_conductivity(self, T, RH=0.99):
        """Calculate membrane ionic conductivity
        
        Args:
            T (float): Temperature (K)
            RH (float): Relative humidity (0-1)
        Returns:
            float: Ionic conductivity (S/m) - Changed from S/cm
        """
        lambda_m = self.water_uptake(RH, T)
        
        # Modified Arrhenius expression for conductivity (adjusted coefficients)
        E_act = 15000  # Activation energy (J/mol)
        sigma_0 = (0.005139 * lambda_m - 0.00326) * 100  # Convert to S/m
        
        return sigma_0 * np.exp(E_act/self.R * (1/303.15 - 1/T))
    
    def electrode_conductivity(self, T):
        """Calculate effective electrode conductivity considering microstructure
        
        Args:
            T (float): Temperature (K)
        Returns:
            float: Effective conductivity (S/m)
        """
        # Temperature dependence of carbon conductivity (adjusted base value)
        sigma_carbon = 10000  # Base conductivity increased and converted to S/m
        
        # Bruggeman correlation for effective conductivity
        return sigma_carbon * (1 - self.porosity)**1.5 / self.tortuosity
    
    def membrane_resistance(self, T, RH=0.99):
        """Calculate area-specific membrane resistance
        
        Args:
            T (float): Temperature (K)
            RH (float): Relative humidity (0-1)
        Returns:
            float: Area-specific resistance (Ω⋅m²)
        """
        return self.t_mem / self.membrane_conductivity(T, RH)
    
    def electrode_resistance(self, T=353.15, P_comp=1.5e6, RH=0.99):
        """Calculate area-specific electrode resistance with enhanced physics
        
        Args:
            T (float): Temperature (K)
            P_comp (float): Compaction pressure (Pa), default 1.5 MPa
            RH (float): Relative humidity (0-1)
        Returns:
            float: Area-specific resistance (Ω⋅m²)
        """
        # 1. Enhanced porosity model with compression effects
        # Based on recent studies on GDL compression (2021-2024)
        P_ref = 1e6  # Reference pressure (1 MPa)
        porosity_0 = self.porosity  # Initial porosity
        c1 = 0.07  # Compression coefficient
        
        # Pressure-dependent porosity
        porosity_compressed = porosity_0 * (1 - c1 * np.log(P_comp/P_ref))
        
        # 2. Temperature and humidity effects on fiber conductivity
        # Base carbon fiber conductivity (S/m)
        sigma_fiber_0 = 60000  # Updated value based on recent measurements
        
        # Temperature effect on fiber conductivity
        alpha_T = -0.0005  # Temperature coefficient (/K)
        T_ref = 298.15  # Reference temperature (K)
        sigma_fiber_T = sigma_fiber_0 * (1 + alpha_T * (T - T_ref))
        
        # 3. Enhanced Bruggeman correlation with anisotropic effects
        # Different exponents for in-plane vs through-plane
        n_ip = 1.5  # In-plane Bruggeman exponent
        n_tp = 2.5  # Through-plane Bruggeman exponent
        
        # Anisotropic effective conductivities
        sigma_eff_ip = sigma_fiber_T * (1 - porosity_compressed)**n_ip
        sigma_eff_tp = sigma_fiber_T * (1 - porosity_compressed)**n_tp / self.tortuosity
        
        # 4. PTFE loading effects
        # Typical PTFE content in modern GDLs
        ptfe_loading = 0.2  # 20% PTFE
        ptfe_factor = 1 - 0.5 * ptfe_loading  # Impact on conductivity
        
        # 5. Contact resistance between fibers
        # Based on percolation theory and recent experimental studies
        R_contact_fibers = 1e-8  # Base contact resistance (Ω⋅m²)
        pressure_exp = 0.7  # Pressure dependence exponent
        R_contact = R_contact_fibers * (P_ref/P_comp)**pressure_exp
        
        # 6. MPL effects (if present)
        mpl_thickness = 30e-6  # MPL thickness (m)
        mpl_porosity = 0.4
        mpl_conductivity = 200  # Base MPL conductivity (S/m)
        
        # MPL effective conductivity
        sigma_mpl = mpl_conductivity * (1 - mpl_porosity)**2
        R_mpl = mpl_thickness / sigma_mpl
        
        # 7. Water effects on contact resistance
        # Based on recent studies of water impact on GDL resistance
        water_factor = 1 + 0.2 * (RH - 0.5)  # Increased resistance at high RH
        
        # 8. Calculate total electrode resistance
        electrode_thickness = 200e-6  # Main GDL thickness (m)
        
        # Through-plane resistance components
        R_bulk_tp = electrode_thickness / sigma_eff_tp
        R_total_single = (R_bulk_tp + R_contact + R_mpl) * water_factor * ptfe_factor
        
        # 9. Crack and deformation effects
        # Based on mechanical degradation studies
        deformation_factor = 1.1  # Accounts for cracks and deformation
        
        # 10. Total resistance for both electrodes
        # Account for anode and cathode differences
        anode_factor = 0.9  # Anode typically has better conductivity
        cathode_factor = 1.1  # Cathode more affected by water/oxidation
        
        R_total = R_total_single * (anode_factor + cathode_factor) * deformation_factor
        
        # 11. Non-uniform compression effects
        # Based on channel/land geometry
        land_ratio = 0.6  # Ratio of land area to total area
        R_under_land = R_total * 0.8  # Better conductivity under lands
        R_under_channel = R_total * 1.3  # Worse under channels
        R_effective = land_ratio * R_under_land + (1 - land_ratio) * R_under_channel
        
        return R_effective
    
    def calculate_interface_resistance(self, P_comp, T):
        """Calculate interfacial resistance between GDL and adjacent layers
        
        Args:
            P_comp (float): Compaction pressure (Pa)
            T (float): Temperature (K)
        Returns:
            float: Interfacial resistance (Ω⋅m²)
        """
        # Interface resistance parameters
        R_interface_0 = 5e-6  # Base interfacial resistance
        pressure_coef = -0.8  # Pressure coefficient
        temp_coef = -0.001   # Temperature coefficient
        
        # Calculate temperature and pressure effects
        T_ref = 298.15
        P_ref = 1e6
        
        return R_interface_0 * (P_comp/P_ref)**pressure_coef * (1 + temp_coef * (T - T_ref))
    
    def bipolar_plate_resistance(self, T=353.15, P_comp=1.5e6):
        """Calculate total bipolar plate resistance including contact resistance with enhanced physics
        
        Args:
            T (float): Temperature (K)
            P_comp (float): Compaction pressure (Pa), default 1.5 MPa
            
        Returns:
            float: Area-specific resistance (Ω⋅m²)
        """
        # 1. Temperature-dependent bulk conductivity
        # Based on graphite composite material studies (2020-2023)
        T_ref = 298.15  # Reference temperature (K)
        alpha_T = -0.001  # Temperature coefficient (/K)
        sigma_T = self.bp_conductivity * (1 + alpha_T * (T - T_ref))
        
        # 2. Bulk resistance calculation with geometric effects
        plate_thickness = 0.002  # Plate thickness (m)
        channel_depth = 0.001  # Flow channel depth (m)
        land_ratio = 0.6  # Ratio of land area to total area
        
        # Effect of flow field geometry on current distribution
        effective_thickness = plate_thickness + (channel_depth * (1 - land_ratio))
        bulk_resistance = effective_thickness / sigma_T
        
        # 3. Enhanced contact resistance model
        # Based on recent studies on interfacial resistance (2021-2024)
        
        # 3.1 Pressure-dependent contact resistance
        # Modified empirical relationship from recent literature
        R_contact_base = self.contact_resistance * 1e-7  # Convert from mΩ⋅cm² to Ω⋅m²
        P_ref = 1e6  # Reference pressure (1 MPa)
        n = 0.8  # Pressure sensitivity exponent
        
        # Pressure effect on contact resistance
        R_contact = R_contact_base * (P_ref/P_comp)**n
        
        # 3.2 Temperature effect on contact resistance
        # Based on thermal expansion and surface deformation studies
        beta_T = -0.0015  # Temperature coefficient for contact resistance (/K)
        R_contact_T = R_contact * (1 + beta_T * (T - T_ref))
        
        # 3.3 Surface roughness and coating effects
        # Typical values for coated plates (if applicable)
        roughness_factor = 1.2  # Increase for rougher surfaces
        coating_thickness = 5e-6  # Typical coating thickness (m)
        coating_conductivity = 1e4  # Coating conductivity (S/m)
        
        # Additional resistance from coating
        R_coating = coating_thickness / (coating_conductivity * land_ratio)
        
        # 4. Interfacial corrosion effect (time-dependent)
        # Simplified model for oxide layer formation
        oxide_thickness = 1e-9  # Base oxide thickness (m)
        oxide_conductivity = 1e-2  # Oxide layer conductivity (S/m)
        R_oxide = oxide_thickness / oxide_conductivity
        
        # 5. Total resistance calculation
        R_total = (bulk_resistance * land_ratio +  # Bulk resistance through lands
                   R_contact_T +                    # Temperature-dependent contact resistance
                   R_coating +                      # Coating resistance
                   R_oxide)                        # Oxide layer resistance
        
        # 6. Correction for non-uniform current distribution
        # Based on current density distribution studies
        current_distribution_factor = 1.1  # Typical value from literature
        
        return R_total * current_distribution_factor
    
    def ohmic_overpotential(self, current_density, T, RH=0.99):
        """Calculate total ohmic overpotential
        
        Args:
            current_density (float): Current density (A/cm²)
            T (float): Temperature (K)
            RH (float): Relative humidity (0-1)
        Returns:
            float: Ohmic overpotential (V)
        """
        # Convert current density from A/cm² to A/m²
        current_density_SI = current_density * 10000  
        
        # Sum all resistance contributions
        R_total = (self.membrane_resistance(T, RH) + 
                  self.electrode_resistance(T) + 
                  self.bipolar_plate_resistance())
        
        return current_density_SI * R_total

class IntegratedOhmicAnalysis(EnhancedOhmicModel):
    """Enhanced analysis class that integrates with the improved physical model"""
    
    def temperature_sensitivity(self, T_range, current_density=1.0, RH=0.99):
        """Analyze sensitivity to temperature with humidity effects"""
        temps = np.linspace(T_range[0], T_range[1], 50)
        overpotentials = []
        resistances = {'membrane': [], 'electrode': [], 'bipolar': []}
        water_contents = []
        
        for T in temps:
            overpotentials.append(self.ohmic_overpotential(current_density, T, RH))
            resistances['membrane'].append(self.membrane_resistance(T, RH))
            resistances['electrode'].append(self.electrode_resistance(T))
            resistances['bipolar'].append(self.bipolar_plate_resistance())
            water_contents.append(self.water_uptake(RH, T))
            
        return temps, overpotentials, resistances, water_contents
    
    def humidity_sensitivity(self, RH_range, T_cell=353.15, current_density=1.0):
        """Analyze sensitivity to relative humidity"""
        humidities = np.linspace(RH_range[0], RH_range[1], 50)
        overpotentials = []
        water_contents = []
        conductivities = []
        
        for RH in humidities:
            overpotentials.append(self.ohmic_overpotential(current_density, T_cell, RH))
            water_contents.append(self.water_uptake(RH, T_cell))
            conductivities.append(self.membrane_conductivity(T_cell, RH))
            
        return humidities, overpotentials, water_contents, conductivities
    
    def current_density_sensitivity(self, i_range, T_cell=353.15, RH=0.99):
        """Analyze sensitivity to current density with humidity effects"""
        currents = np.linspace(i_range[0], i_range[1], 50)
        overpotentials = []
        
        for i in currents:
            overpotentials.append(self.ohmic_overpotential(i, T_cell, RH))
            
        return currents, overpotentials
    
    def membrane_thickness_sensitivity(self, thickness_range, T_cell=353.15, current_density=1.0, RH=0.99):
        """Analyze sensitivity to membrane thickness with humidity effects"""
        thicknesses = np.linspace(thickness_range[0], thickness_range[1], 50)
        overpotentials = []
        original_thickness = self.t_mem
        
        for t in thicknesses:
            self.t_mem = t
            overpotentials.append(self.ohmic_overpotential(current_density, T_cell, RH))
            
        self.t_mem = original_thickness
        return thicknesses, overpotentials
    
    def plot_enhanced_analysis(self):
        """Create comprehensive analysis plots including humidity effects"""
        fig = plt.figure(figsize=(15, 12))
        gs = GridSpec(3, 2, figure=fig)
        
        # Temperature analysis with humidity effects
        temps, temp_overpotentials, resistances, water_contents = self.temperature_sensitivity((293.15, 363.15))
        
        # Plot 1: Temperature effect on overpotential
        ax1 = fig.add_subplot(gs[0, 0])
        ax1.plot(temps - 273.15, temp_overpotentials)
        ax1.set_xlabel('Temperature (°C)')
        ax1.set_ylabel('Ohmic Overpotential (V)')
        ax1.set_title('Temperature Effect on Overpotential')
        ax1.grid(True)
        
        # Plot 2: Resistance components
        ax2 = fig.add_subplot(gs[0, 1])
        ax2.plot(temps - 273.15, resistances['membrane'], label='Membrane')
        ax2.plot(temps - 273.15, resistances['electrode'], label='Electrode')
        ax2.plot(temps - 273.15, resistances['bipolar'], label='Bipolar Plate')
        ax2.set_xlabel('Temperature (°C)')
        ax2.set_ylabel('Resistance (Ω⋅cm²)')
        ax2.set_title('Temperature Effect on Resistance Components')
        ax2.legend()
        ax2.grid(True)
        
        # Plot 3: Humidity analysis
        humidities, RH_overpotentials, water_contents, conductivities = self.humidity_sensitivity((0.3, 0.99))
        ax3 = fig.add_subplot(gs[1, 0])
        ax3.plot(humidities * 100, RH_overpotentials)
        ax3.set_xlabel('Relative Humidity (%)')
        ax3.set_ylabel('Ohmic Overpotential (V)')
        ax3.set_title('Humidity Effect on Overpotential')
        ax3.grid(True)
        
        # Plot 4: Water uptake
        ax4 = fig.add_subplot(gs[1, 1])
        ax4.plot(humidities * 100, water_contents)
        ax4.set_xlabel('Relative Humidity (%)')
        ax4.set_ylabel('Water Content (λ)')
        ax4.set_title('Membrane Water Uptake vs Humidity')
        ax4.grid(True)
        
        # Plot 5: Current density analysis
        currents, current_overpotentials = self.current_density_sensitivity((0, 2.0))
        ax5 = fig.add_subplot(gs[2, 0])
        ax5.plot(currents, current_overpotentials)
        ax5.set_xlabel('Current Density (A/cm²)')
        ax5.set_ylabel('Ohmic Overpotential (V)')
        ax5.set_title('Current Density Effect on Overpotential')
        ax5.grid(True)
        
        # Plot 6: Membrane conductivity
        ax6 = fig.add_subplot(gs[2, 1])
        ax6.plot(humidities * 100, conductivities)
        ax6.set_xlabel('Relative Humidity (%)')
        ax6.set_ylabel('Membrane Conductivity (S/cm)')
        ax6.set_title('Membrane Conductivity vs Humidity')
        ax6.grid(True)
        
        plt.tight_layout()
        return fig
    



if __name__ == "__main__":
    # Create analysis object
    analysis = IntegratedOhmicAnalysis()
    
    # Generate and save enhanced analysis plots
    fig_2d = analysis.plot_enhanced_analysis()
    fig_2d.savefig('enhanced_ohmic_analysis_2d.png')
    
    # Print comprehensive sensitivity analysis
    T_ref = 353.15  # 80°C
    i_ref = 1.0     # A/cm²
    RH_ref = 0.8    # 80% RH
    
    print("\nEnhanced Sensitivity Analysis at Reference Conditions:")
    print(f"Reference Temperature: {T_ref-273.15:.1f}°C")
    print(f"Reference Current Density: {i_ref:.1f} A/cm²")
    print(f"Reference Relative Humidity: {RH_ref*100:.1f}%")
    
    # Temperature sensitivity
    dT = 5  # Temperature perturbation in K
    eta_T_plus = analysis.ohmic_overpotential(i_ref, T_ref + dT, RH_ref)
    eta_T_minus = analysis.ohmic_overpotential(i_ref, T_ref - dT, RH_ref)
    deta_dT = (eta_T_plus - eta_T_minus) / (2 * dT)
    print(f"\nTemperature Sensitivity: {deta_dT:.6f} V/K")
    
    # Current density sensitivity
    di = 0.1  # Current density perturbation in A/cm²
    eta_i_plus = analysis.ohmic_overpotential(i_ref + di, T_ref, RH_ref)
    eta_i_minus = analysis.ohmic_overpotential(i_ref - di, T_ref, RH_ref)
    deta_di = (eta_i_plus - eta_i_minus) / (2 * di)
    print(f"Current Density Sensitivity: {deta_di:.6f} V/(A/cm²)")
    
    # Humidity sensitivity
    dRH = 0.05  # RH perturbation
    eta_RH_plus = analysis.ohmic_overpotential(i_ref, T_ref, RH_ref + dRH)
    eta_RH_minus = analysis.ohmic_overpotential(i_ref, T_ref, RH_ref - dRH)
    deta_dRH = (eta_RH_plus - eta_RH_minus) / (2 * dRH)
    print(f"Humidity Sensitivity: {deta_dRH:.6f} V/RH")
    
    plt.show()



import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

class ParameterInteractionAnalysis(IntegratedOhmicAnalysis):
    def __init__(self):
        super().__init__()
        
    def generate_parameter_space(self):
        """Generate parameter spaces for analysis"""
        # Temperature range: 20°C to 90°C
        temps = np.linspace(293.15, 363.15, 30)
        
        # Current density range: 0 to 2 A/cm²
        currents = np.linspace(0, 2, 30)
        
        # Relative humidity range: 30% to 100%
        humidities = np.linspace(0.3, 1.0, 30)
        
        return temps, currents, humidities
    
    def calculate_interaction_matrices(self):
        """Calculate interaction matrices for all parameter combinations"""
        temps, currents, humidities = self.generate_parameter_space()
        
        # Initialize matrices for each parameter combination
        temp_current_matrix = np.zeros((len(temps), len(currents)))
        temp_humidity_matrix = np.zeros((len(temps), len(humidities)))
        current_humidity_matrix = np.zeros((len(currents), len(humidities)))
        
        # Calculate overpotential for each combination
        for i, T in enumerate(temps):
            for j, i_den in enumerate(currents):
                temp_current_matrix[i,j] = self.ohmic_overpotential(i_den, T)
                
        for i, T in enumerate(temps):
            for j, RH in enumerate(humidities):
                temp_humidity_matrix[i,j] = self.ohmic_overpotential(1.0, T, RH)
                
        for i, i_den in enumerate(currents):
            for j, RH in enumerate(humidities):
                current_humidity_matrix[i,j] = self.ohmic_overpotential(i_den, 353.15, RH)
                
        return {
            'temp_current': (temps, currents, temp_current_matrix),
            'temp_humidity': (temps, humidities, temp_humidity_matrix),
            'current_humidity': (currents, humidities, current_humidity_matrix)
        }
    
    def plot_parameter_interactions(self):
        """Create comprehensive visualization of parameter interactions"""
        matrices = self.calculate_interaction_matrices()
        
        # Create figure with subplots
        fig = plt.figure(figsize=(20, 15))
        
        # 1. Temperature-Current Density Interaction
        temps, currents, temp_current_matrix = matrices['temp_current']
        
        # Contour plot
        ax1 = fig.add_subplot(231)
        X, Y = np.meshgrid(currents, temps-273.15)
        contour = ax1.contour(X, Y, temp_current_matrix, levels=15)
        ax1.clabel(contour, inline=True, fontsize=8)
        ax1.set_xlabel('Current Density (A/cm²)')
        ax1.set_ylabel('Temperature (°C)')
        ax1.set_title('Temperature-Current Density\nOverpotential Contours (V)')
        
        # 3D surface plot
        ax2 = fig.add_subplot(232, projection='3d')
        surf = ax2.plot_surface(X, Y, temp_current_matrix, cmap=cm.viridis)
        ax2.set_xlabel('Current Density (A/cm²)')
        ax2.set_ylabel('Temperature (°C)')
        ax2.set_zlabel('Overpotential (V)')
        ax2.set_title('Temperature-Current Density\n3D Interaction')
        
        # 2. Temperature-Humidity Interaction
        temps, humidities, temp_humidity_matrix = matrices['temp_humidity']
        
        # Contour plot
        ax3 = fig.add_subplot(233)
        X, Y = np.meshgrid(humidities*100, temps-273.15)
        contour = ax3.contour(X, Y, temp_humidity_matrix, levels=15)
        ax3.clabel(contour, inline=True, fontsize=8)
        ax3.set_xlabel('Relative Humidity (%)')
        ax3.set_ylabel('Temperature (°C)')
        ax3.set_title('Temperature-Humidity\nOverpotential Contours (V)')
        
        # 3D surface plot
        ax4 = fig.add_subplot(234, projection='3d')
        surf = ax4.plot_surface(X, Y, temp_humidity_matrix, cmap=cm.viridis)
        ax4.set_xlabel('Relative Humidity (%)')
        ax4.set_ylabel('Temperature (°C)')
        ax4.set_zlabel('Overpotential (V)')
        ax4.set_title('Temperature-Humidity\n3D Interaction')
        
        # 3. Current Density-Humidity Interaction
        currents, humidities, current_humidity_matrix = matrices['current_humidity']
        
        # Contour plot
        ax5 = fig.add_subplot(235)
        X, Y = np.meshgrid(humidities*100, currents)
        contour = ax5.contour(X, Y, current_humidity_matrix, levels=15)
        ax5.clabel(contour, inline=True, fontsize=8)
        ax5.set_xlabel('Relative Humidity (%)')
        ax5.set_ylabel('Current Density (A/cm²)')
        ax5.set_title('Current Density-Humidity\nOverpotential Contours (V)')
        
        # 3D surface plot
        ax6 = fig.add_subplot(236, projection='3d')
        surf = ax6.plot_surface(X, Y, current_humidity_matrix, cmap=cm.viridis)
        ax6.set_xlabel('Relative Humidity (%)')
        ax6.set_ylabel('Current Density (A/cm²)')
        ax6.set_zlabel('Overpotential (V)')
        ax6.set_title('Current Density-Humidity\n3D Interaction')
        
        plt.tight_layout()
        return fig
    
    def calculate_interaction_metrics(self):
        """Calculate key metrics for parameter interactions"""
        matrices = self.calculate_interaction_matrices()
        
        metrics = {}
        
        # Calculate sensitivity metrics for each parameter pair
        for interaction, (param1, param2, matrix) in matrices.items():
            # Calculate average gradient along each parameter
            grad1 = np.gradient(matrix, param1, axis=0)
            grad2 = np.gradient(matrix, param2, axis=1)
            
            # Calculate mean sensitivities
            mean_sensitivity1 = np.mean(np.abs(grad1))
            mean_sensitivity2 = np.mean(np.abs(grad2))
            
            # Calculate interaction strength (correlation between gradients)
            interaction_strength = np.corrcoef(grad1.flatten(), grad2.flatten())[0,1]
            
            metrics[interaction] = {
                'mean_sensitivity1': mean_sensitivity1,
                'mean_sensitivity2': mean_sensitivity2,
                'interaction_strength': interaction_strength
            }
            
        return metrics

if __name__ == "__main__":
    # Create analysis object
    analysis = ParameterInteractionAnalysis()
    
    # Generate and save interaction plots
    fig = analysis.plot_parameter_interactions()
    fig.savefig('parameter_interactions.png', dpi=300, bbox_inches='tight')
    
    # Calculate and print interaction metrics
    metrics = analysis.calculate_interaction_metrics()
    
    print("\nParameter Interaction Analysis:")
    print("\nTemperature-Current Density Interaction:")
    print(f"Temperature Sensitivity: {metrics['temp_current']['mean_sensitivity1']:.6f} V/K")
    print(f"Current Density Sensitivity: {metrics['temp_current']['mean_sensitivity2']:.6f} V/(A/cm²)")
    print(f"Interaction Strength: {metrics['temp_current']['interaction_strength']:.6f}")
    
    print("\nTemperature-Humidity Interaction:")
    print(f"Temperature Sensitivity: {metrics['temp_humidity']['mean_sensitivity1']:.6f} V/K")
    print(f"Humidity Sensitivity: {metrics['temp_humidity']['mean_sensitivity2']:.6f} V/RH")
    print(f"Interaction Strength: {metrics['temp_humidity']['interaction_strength']:.6f}")
    
    print("\nCurrent Density-Humidity Interaction:")
    print(f"Current Density Sensitivity: {metrics['current_humidity']['mean_sensitivity1']:.6f} V/(A/cm²)")
    print(f"Humidity Sensitivity: {metrics['current_humidity']['mean_sensitivity2']:.6f} V/RH")
    print(f"Interaction Strength: {metrics['current_humidity']['interaction_strength']:.6f}")
    
    plt.show()


analysis = ParameterInteractionAnalysis()
fig = analysis.plot_parameter_interactions()
metrics = analysis.calculate_interaction_metrics()
```
